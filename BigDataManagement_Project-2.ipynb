```python
#Transformation on top of the raw data extracted - 
#WE are interested in objects - creating albumn table - we need above information and store it by creating list of it
#on single item - we can iterate over data item

'''for row in data['items']:
    data['items'][0]['track']['album']['id']
    data['items'][0]['track']['album']['name']
    data['items'][0]['track']['album']['total_tracks']
    data['items'][0]['track']['album']['release_date']'''

album_list = []
for row in data['items']:
    album_id = row['track']['album']['id']
    album_name = row['track']['album']['name']
    album_release_date = row['track']['album']['release_date']
    album_total_tracks = row['track']['album']['total_tracks']
    album_url = row['track']['album']['external_urls']['spotify']
    #print(album_name) and creating a dictionary for all the info we have, cleaned json to what we need
    #we have to convert the ouotput into dictionary, so we can create DataFrame oout of it. 
    #lets store to entire 50 records - by creating lists out of it - append element inside album element
    album_element = {'album_id' : album_id,
                    'name' : album_name,
                    'release_date' : album_release_date,
                    'total_tracks' : album_total_tracks,
                    'url' : album_url}
    #print(album_element)
    album_list.append(album_element)
#we have structured the data for the ablum, we can do it for the tracks. 
```

```python
artist_list = []
for row in data['items']:
    for key, value in row.items():
        if key == 'track':
            for artist in value['artists']: #this for loop will provide us the output in dictionary.
                artist_dict = {'artist_id' : artist['id'],
                'artist_name' : artist['name'],
                'external_url' : artist['href'],
                }
                artist_list.append(artist_dict)
```

```python
#similarly we shall create song table, along
song_list = []
for row in data['items']:
    song_id = row['track']['id']
    song_name = row['track']['name']
    song_duration = row['track']['duration_ms']
    song_url = row['track']['external_urls']['spotify']
    song_popularity = row['track']['popularity']
    song_added = row['added_at']
    album_id = row['track']['album']['id']
    artist_id = row['track']['album']['artists'][0]['id']
    song_element = {'song_id':song_id,
                    'song_name':song_name,
                    'duration_ms':song_duration,
                    'url':song_url,
                    'popularity':song_popularity,
                    'song_added':song_added,
                    'album_id':album_id,
                    'artist_id':artist_id
                   }
    song_list.append(song_element)
```
